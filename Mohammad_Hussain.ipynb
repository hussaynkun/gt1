```python
#10
#Ans
def add(*num):
    sum = 0
    for i in num:
        sum = sum+i
    return sum


def sub(*num):
    sum = 0
    for i in num:
        sum = sum-i
    return sum

def mul(b):
    total = 1
    for i in b:
        total = total *i
    return total
```
